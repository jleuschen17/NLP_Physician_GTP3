```python
import pandas as pd
import os
```

```python
dfs = {}
```

```python
directory = 'Mimic II Test'
for file in os.listdir(directory):
    f = os.path.join(directory, file)
    # checking if it is a file and yeah
    if os.path.isfile(f):
        dfs[file] = pd.read_csv(f)
```

```python
for df in dfs:
    print(df)
    print("---------------------------------------------")
    print(dfs[df].iloc[:5])
```
